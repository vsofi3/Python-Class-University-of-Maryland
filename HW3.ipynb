#### Instructions (4 points)
- Write an expression to select each of the following from the Europe data:

   - GDP per capita for all countries in 1982.
    - GDP per capita for Denmark for all years.
    - GDP per capita for all countries for years after 1985.
    - Find the growth of GDP per capita for each country between 2007 and 1952 (i.e., 2007 value divided by 1952 value). Store the result in a new column named `"GDP_pc_growth"`.

````python
# GDP per capita for all countries in 1982.
print(europe_df.loc[:, 'gdpPercap_1982'])
````

````python
# GDP per capita for Denmark for all years.
print(europe_df.loc['Denmark', :])
````

````python
# GDP per capita for all countries for years *after* 1985.
print(europe_df.loc[:,'gdpPercap_1985':])
````

````python
# GDP per capita for each country in 2007 as a multiple of GDP per capita for that country in 1952. Store the result in a new column named "GDP_pc_growth"
europe_df['GDP_pc_growth'] = europe_df.loc[:, 'gdpPercap_2007'] / europe_df.loc[:, 'gdpPercap_1952']
print(europe_df['GDP_pc_growth'])
````

### 3. pandas hist, pdf and cdf

Your job is to plot a PDF and CDF for the fraction column of the tips dataset. This column contains information about what ```fraction``` of the total bill is comprised of the tip.

Remember, when plotting the PDF, you need to specify ```density=True``` in your call to ```.hist()```, and when plotting the CDF, you need to specify ```cumulative=True``` in addition to ```density=True```.

All necessary modules have been imported and the tips dataset is available in the workspace as ```df```. 

````python
data_file = 'https://github.com/huangpen77/BUDT704/raw/main/Chapter06/tips.csv'
df = pd.read_csv(data_file)
df.head()
````

#### Instructions (2 points)

* Plot a PDF for the values in `fraction` with 30 `bins` and a range between 0 and 30%.  

````python
# Plot the PDF
df.fraction.plot(kind='hist', bins=30, density=True, range=(0,0.3))
````

#### Instructions (2 points)
 
* Plot a CDF for the values in `fraction` with 30 `bins` and a range between 0 and 30%. 

````python
# Plot the CDF
df.fraction.plot(kind='hist', bins=30, density=True, cumulative=True, range=(0,0.3))
````
