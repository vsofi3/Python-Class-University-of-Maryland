#  Q-1 The Best Striker

**The Best attacking striker for the league can be judged on the parameter of how many goal contribution that particular player has**<br>
We will use the premier league data of 2019 to determine the best goal-scorer/striker. First we will write the code to determine the top goal scorers of the league 

```python
Best_Attackers = Football_stats[Football_stats['position']=='Forward'].sort_values('goals', ascending=False)[0:4]
Best_Attackers.reset_index(drop=True, inplace=True)   
Best_Attackers

plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = (7,5)
fig, ax = plt.subplots()

ax.bar(Best_Attackers['full_name'], Best_Attackers['goals'])
ax.set_title('TOP Goal Scorers')
ax.set_ylabel('Number of Goals')
```

# THE BEST STRIKER DILEMMA

By the analysis of the bar chart above there are three players with the same number goals and the 4th player has just one goal lesser than the top 3. To determine who was the most prolific amongst the 4 players and who was the best attacker in the league we will analyze the data on the basis of 4 parameters that we have choosen  
* Number of appearances(higher the better as high number of appearances directly correlates with optimal fitness levels)
* Discipline(Whether the player got any yellow or red cards)
* Number of assists given 
* Minutes played(higher the better as high number of appearances directly correlates with optimal fitness levels)

```python
plt.style.use("ggplot")
fig, ax = plt.subplots(2,2 ,figsize=(20, 14))

ax[0,0].plot(Best_Attackers['full_name'], Best_Attackers['appearances'], color='blue')
ax[0,0].set_title('Appearances made')
ax[0,1].bar(Best_Attackers['full_name'], Best_Attackers['yellow_cards'],color='yellow')
ax[0,1].set_title('Bookings[Yellow_Cards]')
ax[1,0].bar(Best_Attackers['full_name'], Best_Attackers['assists'],color='red')
ax[1,0].set_title('Assists_Given')
ax[1,1].scatter(Best_Attackers['full_name'], Best_Attackers['minutes_played'],color='black')
ax[1,1].set_title('Minutes played')
plt.show()
```

### By the following charts and tables we can analyze that <br>
* The most appearances was made by the  Mohammed Salah of Liverpool Fc<br>
* The most undisciplined player was Sergio Aguero for the season 2018-19 with 3 yellow cards<br>
* The highest number of assisted amongst the top 4 goal scorers was shared by Mohammed Salah and by Sergio Aguero with 8 assists made<br>
* The highest minutes played was also capped off by Mohammed Salah with a total of 3086 mins played.
#### By the parameters we chose and the operations perfomed we can say that Mohammed Salah was the best Strker of the PL in 2018/19

## Now we will take our analysis further by determining the best defensive player of the league 

### The parameters choosen for determining the top defenders is the Number of clean sheets


```python
Best_Defender = Football_stats[Football_stats['position']=='Defender'].sort_values('clean_sheets', ascending=False)[0:4]
Best_Defender.reset_index(drop=True, inplace=True)   
Best_Defender

plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = (7,5)
fig, ax = plt.subplots()

ax.bar(Best_Defender['full_name'], Best_Defender['clean_sheets'])
ax.set_title('Top Defenders')
ax.set_ylabel('Number of clean sheets')
```
