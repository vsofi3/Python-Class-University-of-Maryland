# For the analysis first, we will determine the number of players in each teams squad

```python
players =Football_stats.pivot_table(index='Current Club',values='full_name', aggfunc='count').reset_index()
players

plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = (20,7)
fig, ax = plt.subplots()
ax.set_xticklabels(players['Current Club'], rotation=90)

ax.bar(players['Current Club'], players['full_name'])
ax.set_title('Team Name')
ax.set_ylabel('Player Roster Depth')
```

# Q1 The Best Striker

**The Best attacking striker for the league can be judged on the parameter of how many goal contribution that particular player has**<br>
We will use the premier league data of 2019 to determine the best goal-scorer/striker. First we will write the code to determine the top goal scorers of the league 

```python
Best_Attackers = Football_stats.sort_values('goals', ascending=False)[0:4]
Best_Attackers.reset_index(drop=True, inplace=True)   
Best_Attackers

plt.style.use("ggplot")
plt.rcParams['figure.figsize'] = (7,5)
fig, ax = plt.subplots()

ax.bar(Best_Attackers['full_name'], Best_Attackers['goals'])
ax.set_title('TOP Goal Scorers')
ax.set_ylabel('Number of Goals')
```

# THE BEST STRIKER DILEMMA

By the analysis of the bar chart above there are three players with the same number goals and the 4th player has just one goal lesser than the top 3. To determine who was the most prolific amongst the 4 players and who was the best attacker in the league we will analyze the data on the basis of 4 parameters that we have choosen (in order of priority)
* Number of assists given
* Minutes played
* Number of appearances
* Discipline (whether the player got any yellow or red cards) 

```python
plt.style.use("ggplot")
fig, ax = plt.subplots(2,2 ,figsize=(20, 14))

ax[0,0].bar(Best_Attackers['full_name'], Best_Attackers['assists'],color='red')
ax[0,0].set_title('Assists_Given')
ax[0,1].scatter(Best_Attackers['full_name'], Best_Attackers['minutes_played'],color='black')
ax[0,1].set_title('Minutes played')
ax[1,0].plot(Best_Attackers['full_name'], Best_Attackers['appearances'], color='blue')
ax[1,0].set_title('Appearances made')
ax[1,1].bar(Best_Attackers['full_name'], Best_Attackers['yellow_cards'],color='yellow')
ax[1,1].set_title('Bookings[Yellow_Cards]')

plt.show()
```

### By the following charts and tables we can analyze that <br>
* The highest number of assisted amongst the top 4 goal scorers was shared by Mohammed Salah and by Sergio Aguero with 8 assists made<br>
* The highest minutes played was also capped off by Mohammed Salah with a total of 3086 mins played.
* The least appearances made and the least amount of minutes played was by Sergio Aguero<br>
* The most undisciplined player was Sergio Aguero for the season 2018-19 with 3 yellow cards<br>
#### By the parameters we chose and the operations perfomed we can say that Mohammed Salah was the best Strker of the PL in 2018/19

# To validate our analysis we will use our own search engine now and check whether our analysis was valid or not 

```python
def search(search_term, language = 'en'):
    url = f'https://{language}.wikipedia.org/w/index.php?search={search_term}&fulltext=1&ns0=1'
    res = ses.get(url)
    results = res.html.find('.mw-search-result')
    out_results = []
    for result in results:
        title = result.find('.mw-search-result-heading', first=True).text
        summary = result.find('.searchresult', first=True).text
        data = result.find('.mw-search-result-data', first=True).text
        url = result.find('.mw-search-result-heading', first=True).absolute_links.pop()
        out_results.append({'title': title, 'summary': summary, 'data': data, 'url': url})
    return out_results

app = Bottle()
```

```python
@app.route('/') # or @route('/login')
def index():
    return open('website/index.html').read()

@app.route('/search')
def index():
    query = request.query['query']
    return template(open('website/search.html').read(), results=search(query))

@app.route('/css/<filename>')
def server_static(filename):
    print('CSS Served')
    return static_file(filename, root='website/css')

@app.route('/js/<filename>')
def server_static(filename):
    print('JS Served')
    return static_file(filename, root='website/js')

@app.route('/images/<filename>')
def server_static(filename):
    print('Image Served')
    return static_file(filename, root='website/images1')

print('Serving on http://localhost:8080')
run(app, host='localhost', port=8080)
```
