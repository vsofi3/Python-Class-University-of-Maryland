### 1. filter() and lambda functions

In this exercise, you will practice anonymously embedding a lambda function with `filter()`. The function `filter()` offers a way to filter out elements from a list that satisfy certain criteria.

Your goal in this exercise is to use `filter()` to create, from an input list of strings, a new list that contains only strings that have more than 6 characters.

#### Instructions (2 points)

- In the `filter()` call, pass a lambda function and the list of strings, `fellowship`. The lambda function should check if the number of characters in a string `member` is greater than 6; use the `len()` function to do this. Assign the resulting filter object to `result`.
- Convert `result` to a list and print out the list.

````python
# Create a list of strings: fellowship
fellowship = ['frodo', 'samwise', 'merry', 'pippin', 'aragorn', 'boromir', 'legolas', 'gimli', 'gandalf']

# Use filter() to apply a lambda function over fellowship: result
result = filter((lambda member: len(member) > 6), fellowship)

# Convert result to a list: result_list
result_list = list(result)

# Print result_list
print(result_list)
````

Another interesting function that you've learned is `zip()`, which takes any number of iterables and returns a `zip` object that is an iterator of tuples. #### Instructions (4 points) 

- Using `zip()` with `list()`, create a *list of tuples* from the three lists `mutants`, `aliases`, and `powers` (in that order) and assign the result to `mutant_list`.

### 4. Using * and zip to 'unzip'

Let's play around with `zip()` a little more. There is no *unzip* function for doing the reverse of what `zip()` does. We can, however, reverse what has been zipped together by using `zip()` with a little help from `*`! `*` unpacks an *iterable* such as a list or a tuple into positional arguments in a function call.

In this exercise, you will use `*` in a call to `zip()` to unpack the tuples produced by `zip()`.

Two tuples of strings, `mutants` and `powers` have been pre-loaded.

#### Instructions (4 points)

- Create a `zip` object by using `zip()` on `mutants` and `powers`, in that order. Assign the result to `z1`.
- Print the tuples in `z1` by unpacking them into positional arguments using the `*` operator in a `print()` call.
- Because the previous `print()` call would have exhausted the elements in `z1`, recreate the `zip` object you defined earlier and assign the result again to `z1`.
- Note `z2` is a zip object created by calling the `zip()` function with elements in `z1` as positional arguments (through the use of `*z1`). The print statement prints out the content of `z2` in a list.
- Because `z1` and `z2` are exhausted in the prevous code, you have to recreate them.
- 'Unzip' the tuples in `z1` by assigning the zip object `z2` to `result1` and `result2`, in that order.
- The last `print()` statements shoud print the output of comparing `result1` to `mutants` and `result2` to `powers`. Check if the unpacked `result1` and `result2` are equivalent to `mutants` and `powers`, respectively.

````python
mutants = ('charles xavier', 'bobby drake', 'kurt wagner', 'max eisenhardt', 'kitty pryde')
powers = ('telepathy', 'thermokinesis', 'teleportation', 'magnetokinesis', 'intangibility')

# Create a zip object from mutants and powers: z1
z1 = zip(mutants, powers)

# Print the tuples in z1 by unpacking with *
print(*z1, '\n')

# Re-create a zip object from mutants and powers: z1
z1 = zip(mutants, powers)
# zip together the elements in z1 by using the * operator in a zip() call: z2
z2 = zip(*z1)
print(list(z2), '\n')

# Re-create a zip object from mutants and powers: z1
z1 = zip(mutants, powers)
# zip together the elements in z1 by using the * operator in a zip() call: z2
z2 = zip(*z1)
# 'Unzip' the tuples in z1 by unpacking z2: result1, result2
result1, result2 = z2
print(result1, '\n')
print(result2, '\n')

# Check if unpacked tuples are equivalent to original tuples
print(result1 == mutants, '\n')
print(result2 == powers)
````

### 6. Boolean Masks with logical operators

#### Instructions (1.5 points)
- create an array `x` using np.arange(), which starts with 0 and ends with 10, with a step value of 0.5.
- create a Boolean mask which is the combination of two conditions: `x > 5` and `x < 7.5`. In numpy, you have to use `np.logical_and()` method and pass the two conditions as arguments to create such a compound boolean mask. Store the result in `mask`.
- print out the value of `mask`.
- use `mask` to retrieve the value in `x` that satisfies the conditions.

````python
import numpy as np
# Create an array using np.arange(), with start of 0 and end of 10, with a step value of 0.5
x = np.arange(0,10,0.5)

# create a Boolean mask which is the combination of two conditions: x > 5 and x < 7.5.
mask = np.logical_and(x>5, x<7.5)

# print out mask
print(mask)

# use mask to select the elements in array x
x[mask]
````

#### Instructions (1.5 points)
- create a Boolean mask which is the combination of two conditions: `x < 2` or `x > 8`. In numpy, you have to use `np.logical_or()` method and pass the two conditions as arguments to create such a compound boolean mask. Store the result in `mask2`.
- print out the value of `mask2`.
- use `mask2` to retrieve the value in `x` that satisfies the conditions.

````python
# create a Boolean mask which is the combination of two conditions: x < 2 or x > 8.
mask2 = np.logical_or(x<2,x>8)

# print out mask2
print(mask2)

# use mask2 to select the elements in array x
x[mask2]
````

### 7. Fancy indexing

#### Instructions (1.5 points)
- create a list of squares for a range of integers between 1 and 16 (inclusive). Use a list comprehension for this. Store the resulting array in squares.
- create an numpy array, squares_arr, from squares
- reshape squares_arr to (4,4)

````python
import numpy as np
# create a list of squares for a range of integers between 1 and 16 (inclusive). Use a list comprehension for this. Store the resulting array in squares.
squares = [x**2 for x in range(1,17)]

# create an numpy array, squares_arr, from squares
squares_arr = np.array(squares)

# reshape squares_arr to (4,4)
squares_arr = squares_arr.reshape(4,4) # reshaping to 4 by 4 matrix
squares_arr
````

#### Instructions (1.5 points)
- use np.where() to find out the indexes of the elements in squares_arr that can be divisible by 3. Store the result in two variables `row_indexes` and `column_indexes`.
- print out `row_indexes` and `column_indexes`
- use `row_indexes` and `column_indexes` to retrieve the elements in squares_arr that can be divisible by 3.

````python
# use np.where() to find out the indexes of the elements in squares_arr that can be divisible by 3. Store the result in two variables `row_indexes` and `column_indexes`.
row_indexes, column_indexes = np.where(squares_arr%3 == 0)

# print out row_indexes and column_indexes
print(row_indexes, column_indexes)

# use row_indexes and column_indexes to retrieve the elements in squares_arr that can be divisible by 3.
squares_arr[row_indexes, column_indexes]
````

### 8. Vectorizing .upper()

There are many situations where you might want to use Python methods and functions on array elements in NumPy. You can always write a `for` loop to do this, but vectorized operations are much faster and more efficient, so consider using `np.vectorize()`! 

You've got an array called `names` which contains first and last names:

```python
names = np.array([["Izzy", "Monica", "Marvin"],
                  ["Weber", "Patel", "Hernandez"]])

```

You'd like to use one of the Python methods you learned on DataCamp, `.upper()`, to make all the letters of every name in the array uppercase. As a reminder, `.upper()` is a **string method**, meaning that it must be called on an instance of a string: `str.upper()`.

Your task is to vectorize this Python method. `numpy` is loaded for you as `np`, and the `names` array is available.

####  Instructions (2 points)

- Create a vectorized function called `vectorized_upper` from the Python `.upper()` string method.
- Apply `vectorized_upper()` to the `names` array and save the resulting array as `uppercase_names`.

````python
import numpy as np
names = np.array([["Izzy", "Monica", "Marvin"], ["Weber", "Patel", "Hernandez"]])

# Vectorize the str.upper() string method
vectorized_upper = np.vectorize(str.upper)

# Apply vectorized_upper to the names array
uppercase_names = vectorized_upper(names)
print(uppercase_names)
````
