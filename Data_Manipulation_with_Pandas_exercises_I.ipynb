### 1. Using apply() to transform a column

The `.apply()` method can be used on a pandas DataFrame to apply an arbitrary Python function to every element. In this exercise you'll take daily weather data in Pittsburgh in 2013 obtained from [Weather Underground](https://www.wunderground.com/history).

A function to convert degrees **Fahrenheit** to degrees **Celsius** has been written for you. Your job is to use the `.apply()` method to perform this conversion on the `'Mean TemperatureF'` and `'Mean Dew PointF'` columns of the `weather` DataFrame.

#### Instructions (3 points)

- Apply the `to_celsius()` function over the `'Mean TemperatureF'` and `'Mean Dew PointF'` columns of the `weather` DataFrame, and store the results in a new dataframe `df_celsius`.
- Change the column labels of `df_celsius` to `'Mean TemperatureC'` and `'Mean Dew PointC'` using the `.columns` attribute.

````python
import pandas as pd
import numpy as np
# import the dataset
weather = pd.read_csv("https://github.com/huangpen77/BUDT704/raw/main/Chapter07/pittsburgh2013.csv", parse_dates = ['Date'])
weather.head()
````

````python
# a function to convert degrees Fahrenheit to degrees Celsius: to_celsius
def to_celsius(F):
    return 5/9*(F - 32)

# Apply the function over 'Mean TemperatureF' and 'Mean Dew PointF': df_celsius
df_celsius = weather[['Mean TemperatureF','Mean Dew PointF']].apply(to_celsius)

# Reassign the column labels of df_celsius
df_celsius.columns = ['Mean TemperatureF', 'Mean Dew PointF']

# Print the output of df_celsius.head()
print(df_celsius.head())
````

### 2. Using .map() with a dictionary

The `.map()` method is used to transform values according to a Python dictionary look-up. In this exercise you'll practice this method while working with the `election` DataFrame, which has been pre-loaded for you.

Your job is to use a dictionary to map the values `'Obama'` and `'Romney'` in the `'winner'` column to the values `'blue'` and `'red'`, and assign the output to the new column `'color'`.

#### Instructions (3 points)

- Create a dictionary with the key:value pairs `'Obama':'blue'` and `'Romney':'red'`.
- Use the `.map()` method on the `'winner'` column using the `red_vs_blue` dictionary you created.

````python
# import the dataset and take a look
# county is index column
election = pd.read_csv("https://github.com/huangpen77/BUDT704/raw/main/Chapter07/pennsylvania2012_turnout.csv", index_col = 'county')
election.head()
````

````python
# Create the dictionary: red_vs_blue
red_vs_blue = {'Obama':'blue', 'Romney':'red'}

# Use the dictionary to map the 'winner' column to the new column: election['color']
election['color'] = election['winner'].map(red_vs_blue)

# Print the output of election.head()
print(election.head())
````

### 1. Changing index of a DataFrame

Indexes of dataframes are immutable objects. This means that if you want to change or modify the index in a DataFrame, then you need to change the whole index. You will do this now, using a list comprehension to create the new index.

A list comprehension is a succinct way to generate a list in one line. For example, the following list comprehension generates a list that contains the cubes of all numbers from 0 to 9:

```python
cubes = [i**3 for i in range(10)]
```

#### Instructions (3 points)

* Create a list `new_idx` with the same elements as in `sales.index`, but with all characters in their upper case. Use a list comprehension to do this.
* Assign `new_idx` to `sales.index`.

````python
import pandas as pd 
sales = pd.read_csv('https://github.com/huangpen77/BUDT704/raw/main/Chapter07/sales.csv', index_col = "month")
sales.head()
````

````python
# Create the list of new indexes: new_idx. Use list comprehension to do this.
new_idx = [item.upper() for item in sales.index]
# see how all the months are now in uppercase

# Assign new_idx to sales.index
sales.index = new_idx

# Print the sales DataFrame
print(sales)
````
